```python
# Dictionary of keywords mapped to responses

faq_data = {
    ("shipping", "delivery", "how long"): "Shipping usually takes 3-5 business days within Pakistan, and 7-10 days for remote areas.",
    ("return", "refund", "exchange"): "You can return any item within 7 days of delivery, as long as it's unused and in original packaging.",
    ("order status", "track", "where is my order"): "You can track your order using the tracking ID sent to your email after checkout.",
    ("payment", "cod", "cash on delivery"): "We accept Cash on Delivery, bank transfer, and major debit/credit cards.",
    ("discount", "coupon", "promo"): "Check our Instagram page for active promo codes - we run seasonal discounts every month.",
    ("contact", "support", "help"): "You can reach our support team at support@yourstore.com or via WhatsApp at 0300-XXXXXXX.",
}

default_response = "I'm not sure about that yet - please contact our support team at support@yourstore.com for more help."
```

```python
def get_response(user_input):
    user_input = user_input.lower()

    for keywords, response in faq_data.items():
        for keyword in keywords:
            if keyword in user_input:
                return response

    return default_response
```

```python
test_questions = [
    "How long does shipping take?",
    "Can I return my order?",
    "Do you accept cash on delivery?",
    "Is there any discount code available?",
    "What is the meaning of life?"
]

for question in test_questions:
    print(f"You: {question}")
    print(f"Bot: {get_response(question)}\n")
```
